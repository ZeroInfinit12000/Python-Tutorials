```python
#Set up
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
#Seaborn datasets
print(sns.get_dataset_names())
```

```python
#Load a seaborn dataset
crash_df=sns.load_dataset("car_crashes")
crash_df.head()
```

```python
#displot Distribution Plots--->Plot univariant distributions 
#%matplotlib widget
#density means to simulate the density function pdf, if you write also cumulative=True, you estimate the cdf 
sns.displot(crash_df["not_distracted"],rug=True,kde=True,stat="density",cumulative=True)

#different kinds. First just kde
sns.displot(crash_df["not_distracted"],rug=True,kind="kde")
#Now, kind="ecdf", which is just the empirical cdf. Basically, is 
# the distribution that assigns probability 1/n to each data point n in the graph.
sns.displot(crash_df["not_distracted"],rug=True,kind="ecdf")


```

```python
#Plot two graphs in displot at one!

sns.displot(x="speeding",y="alcohol",data=crash_df,kind="kde",stat="density")
```

```python
#Joing Plot
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="kde",fill=True,thresh=0,cmap='Blues')
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="reg")
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="hex")
```

```python
#Kde plot
sns.kdeplot(crash_df['alcohol']) #Basically, estimates a pdf. Is a nonparametric way of doing this
#if put option cumulative=True, it estimates a cdf
```

```python
#Pair Plots
sns.pairplot(crash_df)
```

```python
#Other PairPlot
tips_df=sns.load_dataset("tips")
tips_df.head()

sns.pairplot(tips_df,diag_kind="auto",hue="sex",palette="Blues")
```

```python
#RugPlot Sort of Histogram
sns.rugplot(tips_df['tip'])
```

# Styling

```python
#Set up
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
crash_df=sns.load_dataset("car_crashes")
#Change figure grid and size
sns.set_style('white')
plt.figure(figsize=(8,4))
#Change x,y labels size
sns.set_context(context="paper",font_scale=1.5)
#plot grid
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="reg")
#Delete axis of left and botton
sns.despine(left=True,bottom=True)
```

```python
# Categorical plots     
tips_df=sns.load_dataset("tips")
sns.barplot(x="sex",y="total_bill",data=tips_df,estimator=np.std,color="r")
```

```python
#Count Plots
sns.countplot(x="sex",data=tips_df)
```

```python
#BoxPlot
sns.boxplot(x="day",y="total_bill",data=tips_df,hue="sex")
#Nota> La linea del centro es la mediana. Los bigotes son creo recordar RIC o alguna especie de derivada
# respectiva a esa cosa...

```

```python
#Violin Plot--> Silimar a BoxPlot pero con kde
sns.violinplot(x="day",y="total_bill",data=tips_df,hue="sex",split=True)

```

```python
#Strip Plot
plt.figure(figsize=(12,7))
sns.stripplot(x="day",y="total_bill",data=tips_df,jitter=True,hue="sex")
```

```python
#Swarm Plot
sns.violinplot(x="day",y="total_bill",data=tips_df,hue="day")
sns.swarmplot(x="day",y="total_bill",data=tips_df,color="white")
```

# Palettes

```python
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.set_context('talk')
sns.stripplot(x="day",y="total_bill",data=tips_df,hue='sex')
#Now, to change the color of the plot, you must set the palette and chose a cmap.

```

```python
sns.stripplot(x="day",y="total_bill",data=tips_df,hue='sex',palette="magma")
#Move legend
plt.legend(loc=0)
```

# Heatmaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
#Create heatmap with data----> Matrix format
crash_df.head()
crash_mx=crash_df.iloc[:,:-1].corr()
crash_mx

sns.heatmap(crash_mx,annot=True,cmap="Blues")
```

```python
#Other way to do it, when we got many strings!
flights=sns.load_dataset("flights")
flights.head()

flights=flights.pivot_table(index="month",columns="year",values="passengers")
flights.head()
sns.heatmap(flights,cmap="Blues",linecolor="white",linewidths="1")
plt.yticks(rotation=45)
```

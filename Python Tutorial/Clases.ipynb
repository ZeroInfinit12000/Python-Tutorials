```python
#Multiple classes interactions

class Student:
    def __init__(self,name,age,grade):
         self.name=name
         self.age=age
         self.grade=grade #0-100
    def  get_grade(self):
        return self.grade 
class Course:
    def __init__(self,name,max_students):
        self.name=name
        self.max_students=max_students
        #new attribute not assigned to arguments
        self.students = []
    def add_student(self,student):
        if len(self.students)<self.max_students:
            self.students.append(student)
            return True
        return False
    def get_averege_grade(self):
        value=0
        for student in self.students:
            value +=student.get_grade()
        return  value/len(self.students)
s1=Student("Tim",19,60)
s2=Student("Bill",19,75)
s3=Student("Eva",19,88)

course=Course("Mathematics",2)
course.add_student(s1)
course.add_student(s2)
print(course.students)
print(course.students[0].name) 
print(course.add_student(s3))
print(course.get_averege_grade())       
```

```python
#Inheritance

#Basically, we got two clases which are basically the same attributes  with just 1 or 2 exceptions not significant, we shall try to normilze this via inheritance.
#In this case only the speak method differes from one class to another, so we shall only keep it and normalize the rest. 
class Cat:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def speak(self):
        print("Meow")
        
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def speak(self):
        print("Bark")
        
#The first step is to remove the things which are the same from both classes, in this case, the __init__ method:
class Cat:
    def speak(self):
        print("Meow")
        
class Dog:
    def speak(self):
        print("Bark")
#Lets make an upper level class, say Pet, which has the attributes of the previous classes.
class Pet:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old")
    def speak(self):
        print("I do not know what sound to make!")
#This attributes and method are what i want my Dog and Cat class to have, so lets add it to them via inheritance. The requirements of the atributes
# to create a instance of the new classes which inherited Pet are the same as creating a instance/object of the more general Pet class, so we
# shall put the correct arguments, as seen here:
#Note: If the parent class have a method defined, say speak, and also the childs have a method name the same, the childs method
#will overwrite the parents method! 
class Cat(Pet):
    def speak(self):
        print("Meow")
        
class Dog(Pet):
    def speak(self):
        print("Bark")
class Fish(Pet):
    pass
p=Pet("Tim",19)
p.show()
p.speak()
c=Cat("Bill",34)
c.show()
c.speak()
d=Dog("Jill",25)
d.show()
d.speak()
f=Fish("Bobby",2)
f.show()
f.speak()


```

```python
#super()

#Now, if we want to give new attributes (or rather, new arguments when creating an object ) to the child class, doing it this way, although
#it might work, it is not recomendable, since we might lose something from the inheritance of the Parent (we overwrite the inhertiance):

class Pet:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old")
    def speak(self):
        print("I do not know what sound to make!")
#Lets create new argument for the __init__ class and new attribute for our cat.
class Cat(Pet):
    def __init__(self,name,age,color):
        self.name=name
        self.age=age
        self.color=color
    def speak(self):
        print("Meow")

#When we do an inheritance like this, the arguments oname, age must somehow refer to the __init__ of our partent class, here Pet. To do this
#we use the super().__init__(name,age), no need for self.
#. Super stands for the class we inherit our attributes for, and basically refers those two arguments to the
#way they were used in the __init__ of the parent class Pet.
class Pet:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old")
    def speak(self):
        print("I do not know what sound to make!")

class Cat(Pet):
    def __init__(self,name,age,color):
        super().__init__(name,age)
        self.color=color
    def speak(self):
        print("Meow")
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old and I have {self.color} color.")
c=Cat("Tim",22,"Brown")
c.show()
```

```python
#Class atributes. 
# There are distinct from object atributes (in all the previous cells, I was refering to object attributes). 
# The main distinction with Object attributes is that they are not defined in any method, and that they do not have self. before them when defining
#since they do not refer to any particular instance of the Class, they are atributes both of the class and to every object of said class.

#Class and even object atributes are easy to change via, say we got attribute x, then just do object.x="new input for x", and for
# class atributes, say y,  classname.y="new class atribute input".
class Person:
    number_of_people=0
   
    def __init__(self,name):
        self.name=name
p1=Person("Tim")
p2=Person("Bob")
print(p1.name)
print(p1.number_of_people)
print(p2.name)
print(p2.number_of_people)
#Change attributes
Person.number_of_people=3
p1.name="Tim 2"
p2.name="Bob 2"
print(p1.name)
print(p1.number_of_people)
print(p2.name)
print(p2.number_of_people)


# This is a standard use of Class atributes:
class Person:
    number_of_people=0
   
    def __init__(self,name):
        self.name=name
        Person.number_of_people+=1
print(Person.number_of_people)
p1=Person("Tim")
print(Person.number_of_people)
p2=Person("Bob")
print(Person.number_of_people)
```

```python
#Class methods
#Basically, are methods that apply to all the class. In general, a nice use they have is to change CLass atributes. Insted of Person.number_of_people+=1
#We might just define a class method. The thing that identifies them is that> 1) they have a decorator @classmethod above the methnod, and 2)
#they have cls as an argument rather than self.
class Person:
    number_of_people=0
    def __init__(self,name):
        self.name=name
        Person.add_person()
    @classmethod
    def number_of_people_(cls):
        return cls.number_of_people
    @classmethod
    def add_person(cls):
        cls.number_of_people +=1
print(Person.number_of_people)
p1=Person("Tim")
print(Person.number_of_people)
p2=Person("Bob")
print(Person.number_of_people)
```

```python
#static methods

#It apperas that they are just your standard functions, but inside of a class as methods. They do not refer to any instance, nor they change
#anything particular to the class. They do not need self or cls arguments when being defined.age

#Suppose you have a as.py document and you got the following functions.

def addition(x:int or float,y:int or float)->int or float:
    print(x+y)
def product(x:int or float,y:int or float)->int or float:
    print(x*y)
def root_n(x:int or float,n:int or float)->int or float:
    print(x**(1/n))
    
#Now, when you import this document, you can always do just import as, and refer to any function with as.addition(x,y)....
#To add all this function in one class we could use static methods
class Math:
    @staticmethod
    def addition(x:int or float,y:int or float)->int or float:
        print(x+y)
    @staticmethod
    def product(x:int or float,y:int or float)->int or float:
        print(x*y)
    @staticmethod
    def root_n(x:int or float,n:int or float)->int or float:
        print(x**(1/n))
#Now we from as.py import Math, and we can use this like this:
Math.addition(2,5)
Math.root_n(2,2)
#or create a instance
math=Math()
math.addition(2,5)
math.root_n(2,2)
```

## Matrix Multiplication

There are two main ways to do multiplications in neural networks and deep learning:

1. Elements-wise multiplication
2. Matrix multiplication/dot product

```python
# Element wise multiplication
print(tensor,"*",tensor)
print(f"Equals",{tensor*tensor})
```

```python
# Matrix multiplication> 1*1+2*2+3*3. In this case, since we got only
# dim 1 tensors, we perform the standard dot product.
torch.matmul(tensor,tensor)
```

```python
#time diference between matmul vs for loop
#For loop
%%time
value=0
for i in range(len(tensor)):
  value += tensor[i]*tensor[i]
print(value)
```

```python
# Pytorch matmul
%%time
tensor @ tensor
```

## One of the most common errors in deep learning: Shape errors

### Rules that matrix multiplication must satisfy:

1. The **inner dimensions** must match:
* "(3,**2**) @ (**3**,2)" will not work
* "(2,**3**) @ (**3**,2)" will work
* "(3,**2**) @ (**2**,3)" will work
2. The resulting matrix has the shape of the **outer dimensions**:
* None
* (2,2)
* (3,3)

```python
# Error
torch.matmul(torch.rand(3,2),torch.rand(3,2))
```

```python
print(torch.matmul(torch.rand(2,3),torch.rand(3,2)).shape)
print(torch.matmul(torch.rand(3,2),torch.rand(2,3)).shape)
```

```python
tensor_A=torch.rand(3,2)
tensor_B=torch.rand(3,2)
print(tensor_A)
print(tensor_B)
```

```python
torch.mm(tensor_A,tensor_B)
```

## Reshaping, stacking, squeezing and usqueezing tensors

* Reshaping - reshape an input tensor to a defined shape. Just multiply all dimensions of the given tensor. Any other tensor of dimensions such that multiplied can generate the same number is legit.
* View - Return a view of an input tensor of a certain shape but keep the same memory as the original tensor.
* Stacking (torch.stack)- combine multiple tensors on a certain direction:
  1. torch.stack: We must specify the dimension of the stack
  2. torch.vstack: Combine tensors on top of each other
  3. torch.hstack: Combine tensors side by side
* Squeeze - remove all "1" dimensions from a tensor. I guess is equivalent to *x in some cases.
* Unsqueeze - add a "1" dimension to a target tensor.
* Permute - Return a view of the input tensor with dimensions permuted (swaped) in a certain way.

```python
# Tensor
x = torch.arange(1.,10.)
x,x.shape
```

```python
# Add an extra dimension
x_reshape=x.reshape(1,9)
x_reshape
```

```python
# Change the view. This is the same as before, the diference here is that
# z shares the memory with x, hence if we where to chage z, we will be chaning
# x as well!
z=x.view(1,9)
z,z.shape
```

```python
z[:,0]=5
z,x
```

```python
# Stack tensors on top of each other
x_stacked=torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
#  Squeeze
x_reshape,x_reshape.shape
```

```python
x_reshape.squeeze(),x_reshape.squeeze().shape
```

```python
# torch.unsqueeze()- adds a single dimension to a target tensor at a specific dim
x_reshape,x_reshape.shape
```

```python
# Just note that dim=0, is 1, dim=2, is 9, hence if we unsqueeze at dim=2, we
# add a new element in dim=3!
x_reshape.unsqueeze(dim=2),x_reshape.unsqueeze(dim=2).shape
```

```python
# permute> Basically change the order of the dimensions!!!
x=torch.rand(2,3,5)
x.size()
```

```python
x.permute(2,0,1).size()
```

```python
# Example usage
x_original=torch.rand(size=(224,224,3)) # [height,width,colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2,0,1) # shifst axis 0->!, 1->2, 2->0
x_original.shape,x_permuted.shape # [colour_channels,height,width]
```

## PyTorch tensors and NumPy

NumPy and PyTorch has interactivity between both!
* Data in NumPy, want it in PyTorch -> torch.from_numpy(ndarray)
* PyTorch tensor -> NumPy -> torch.Tensor.numpy()

```python
# Numpy to tensor
import torch
import numpy as np
array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array) # Warning, Numpy datatype is float64, but torch is float32!
array, tensor
```

```python
# Change the value of the array, what about the tensor?-> Nothing!
array = array+1
array,tensor
```

```python
# Tensor to NumPy
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor?-> Nothing
tensor  = tensor+1
tensor, numpy_tensor
```

```python
#Multinominal Logit

#La interpretacion es basicamente Odd Ratio de cada uno de las variables. Si haces Odd Ratio te sale facil...
#No hay cuts aca
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy.stats import bootstrap
#Load
data = pd.read_csv(r"C:\Users\zeroi\Desktop\Programing\Econometria\Datos\keane_P42.csv")
data=data.dropna()
#Variables
Y=data["status"]
X=data.iloc[:,1:]
X=sm.add_constant(X)
# Fit MNLogit model
mnlogit_mod = sm.MNLogit(Y, X).fit()
print(mnlogit_mod.summary())

#Bootstrap
n_obs=len(data)
n_boot=4
b_educ_boot=[]
for i in range(n_boot):
    data=data.sample(n=n_obs,replace=True,random_state=11)
    Y=data["status"]
    X=sm.add_constant(data.iloc[:,1:])
    mnlogit_mod = sm.MNLogit(Y, X).fit(disp=0)
    coefficients = mnlogit_mod.params.iloc[1]
    b_educ_boot.append(coefficients)
b_educ_boot=np.array(b_educ_boot)
b_educ_mean=np.mean(b_educ_boot,axis=0)
b_educ_std=np.std(b_educ_boot,axis=0)
print("b_educ_mean bootstraps:",b_educ_mean)
print("b_educ_std bootstraps:",b_educ_std)
```

```python
import numpy as np
import statsmodels.api as sm
from scipy.stats import bootstrap

# Generate some example data
np.random.seed(42)
X = np.random.rand(100, 1)
y = 3 * X.squeeze() + 2 + np.random.normal(0, 1, 100)

# Fit the original OLS model
X_with_intercept = sm.add_constant(X)
model = sm.OLS(y, X_with_intercept)
original_results = model.fit()

# Number of bootstrap samples
num_bootstraps = 1000

# Bootstrap resampling for regression parameters with statsmodels OLS and scipy.stats.bootstrap
def fit_model(data, _):
    X_bootstrap = data[:, 0]
    y_bootstrap = data[:, 1]

    X_bootstrap_with_intercept = sm.add_constant(X_bootstrap)
    model_bootstrap = sm.OLS(y_bootstrap, X_bootstrap_with_intercept)
    results_bootstrap = model_bootstrap.fit()

    return results_bootstrap.params

# Perform bootstrap using scipy.stats.bootstrap
bootstrapped_params = bootstrap(np.column_stack((X, y)), statistic=fit_model, num_iterations=num_bootstraps, num_samples=len(X))

# Calculate 95% confidence intervals for intercept and slope
confidence_intervals = np.percentile(bootstrapped_params, [2.5, 97.5], axis=0)

print("Original Coefficients:")
print("Intercept:", original_results.params[0])
print("Slope:", original_results.params[1])

print("\nBootstrap Coefficients 95% CI:")
print("Intercept:", confidence_intervals[0])
print("Slope:", confidence_intervals[1])
```

```python
import numpy as np
import statsmodels.api as sm

# Generate some example data
np.random.seed(333)
X = np.random.rand(100, 1)
y = 3 * X.squeeze() + 2 + np.random.normal(0, 1, 100)

# Fit the original OLS model
X_with_intercept = sm.add_constant(X)
model = sm.OLS(y, X_with_intercept)
original_results = model.fit()

# Number of bootstrap samples
num_bootstraps = 10000

# Bootstrap resampling for regression parameters with statsmodels OLS
bootstrap_coefs = np.zeros((num_bootstraps, 2))
for i in range(num_bootstraps):
    indices = np.random.choice(len(X), len(X), replace=True)
    X_bootstrap = X[indices]
    y_bootstrap = y[indices]
    
    X_bootstrap_with_intercept = sm.add_constant(X_bootstrap)
    model_bootstrap = sm.OLS(y_bootstrap, X_bootstrap_with_intercept)
    results_bootstrap = model_bootstrap.fit()
    
    bootstrap_coefs[i] = results_bootstrap.params

# Calculate 95% confidence intervals for intercept and slope
confidence_intervals = np.percentile(bootstrap_coefs, [2.5, 97.5], axis=0)

print("Original Coefficients:")
print("Intercept:", original_results.params[0])
print("Slope:", original_results.params[1])

print("\nBootstrap Coefficients 95% CI:")
print("Intercept:", confidence_intervals[0])
print("Slope:", confidence_intervals[1])
```

```python
import numpy as np
from scipy.stats import bootstrap
import statsmodels.api as sm
# Generate some example data
np.random.seed(42)
data = np.random.normal(loc=5, scale=2, size=(100,2))  # Ensure data is a 2D array
x=data[:,0]
y=data[:,1]
def ols(data):
    x=data[:,0]
    y=data[:,1]
    model=sm.OLS(y,x).fit()
    coeff=model.params[0]
    return coeff

model=sm.OLS(y,x).fit()
print(model.summary())
# Use bootstrap to estimate the mean and its confidence interval
result = bootstrap((data,), ols)
print(result.bootstrap_distribution)
# Print the results
print("Bootstrap Mean:", result.standard_error)
print("Bootstrap 95% Confidence Interval:", result.confidence_interval)
```
